```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
print(model)
```

```python
X = df_salary[['YearsExperience']]
Y = df_salary[['Salary']]
```

```python
train_model=model.fit(X,Y)
train_model
```

```python
train_model.coef_
```

```python
train_model.intercept_
```

```python
y_pred = train_model.predict(X)
y_pred
y_predict = pd.DataFrame(y_pred)
```

```python
from sklearn.metrics import r2_score
accurracy = r2_score(Y,y_pred)*100
print (accurracy)
```

```python
newcomparision = pd.DataFrame({'Actual Salary':Y.iloc[:,0],'Predicted Salary':y_predict.iloc[:,0]})
print(newcomparision)
```
