```python
X_encoded = cardetails_data.drop('Price',axis=1)
y = cardetails_data['Price']
X = pd.get_dummies(X_encoded, drop_first=True)
print(X.columns)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size= 0.2)
```

```python
X_train.shape,y_train.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
price_prediction_model = LinearRegression()
print(price_prediction_model)
```

```python
price_prediction_model.fit(X,y)
```

```python
predict_price = price_prediction_model.predict(X_test)
predict_price
```

```python
price_prediction_model.coef_
```

```python
price_prediction_model.intercept_
```

```python
pd.DataFrame({"Actual_carprice":y_test,"Predicted_carprice":predict_price})
```

```python
from sklearn.metrics import r2_score
accurracy = r2_score(y_test,predict_price)*100
print("Accurracy of Price prediction model:",accurracy)
```

```python
Error_model = 100 - accurracy
print("Error of model accurracy:",Error_model)
```
