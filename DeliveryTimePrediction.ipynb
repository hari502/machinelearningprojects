```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
print(model)
```

```python
y = data[['Delivery Time']]
X = data[['Sorting Time']]
```

```python
train_model = model.fit(X,y)
train_model
```

```python
train_model.coef_
```

```python
train_model.intercept_
```

```python
##y = (1.6490199*9)+6.58273397
##print(y)
```

```python
y_predict = train_model.predict(X)
print(y_predict)
```

```python
Y_pred = pd.DataFrame(y_predict)
```

```python
yvsnewy = pd.DataFrame({'Predicted':Y_pred.iloc[:,0], 'Actual':y.iloc[:,0]})
print(yvsnewy)
```

```python
from sklearn.metrics import r2_score
Accuracy = r2_score(y, y_predict)*100
print("Accuracy of model-",Accuracy)
```
